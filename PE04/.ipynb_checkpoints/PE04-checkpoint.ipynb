```python
# PE04 is an exercise related to pandas. Please follow the steps below.

# Generate a DataFrame which contains 48 random integers 
# between 0 and 49 in 6 rows and 8 columns.
# Set the seed value to 1234 (i.e., np.random.seed(1234)), 
# and use np.random.randint() to generate controlled random numbers. 
# Name the DataFrame data1.
# Use the list of the following alphabets [‘A’, ‘B’, ‘C’, ‘D’, ‘E’, ‘F’] 
# as indexes and the following list 
# ['One', 'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight'] 
# as columns in data1.
# For each row, calculate the interquartile range (IQR) 
# which is the distance between the first and third quartiles. 
# Add it as a new column named ‘IQR’ next to ‘Eight’ in data1.
# Calculate the coefficient of variation (CV = standard deviation / mean) 
# for each row and update data1 by adding it as a new column named ‘CV’.
# Calculate the mean of each column, and add it as a new row named ‘mean’.
# Print data1.

```

```python
import pandas as pd 
import numpy as np
```

```python
np.random.seed(1234)
```

```python
#generate dataframe 'data1'
data1 = pd.DataFrame(np.random.randint(0,50,size=(6, 8)), index= list('ABCDEF'), columns=('One', 'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight'))
data1
```

```python
#calculate the interquartile range (IQR) for each row 

IQR = {'IQR':[
    (data1.iloc[0].quantile(0.75) - data1.iloc[0].quantile(0.25)),
    (data1.iloc[1].quantile(0.75) - data1.iloc[1].quantile(0.25)),
    (data1.iloc[2].quantile(0.75) - data1.iloc[2].quantile(0.25)),
    (data1.iloc[3].quantile(0.75) - data1.iloc[3].quantile(0.25)),
    (data1.iloc[4].quantile(0.75) - data1.iloc[4].quantile(0.25)),
    (data1.iloc[5].quantile(0.75) - data1.iloc[5].quantile(0.25))]}

IQRdf=pd.DataFrame(IQR, columns=['IQR'])
IQRdf.index=list('ABCEDF')

IQRdf
    
```

```python
#clean up indeces and columns names
data1.reset_index(drop=True, inplace=True)
IQRdf.reset_index(drop=True, inplace=True)
```

```python
#add IQR results into data1 dataframe as a new column
data1=pd.concat([data1, IQRdf], axis=1)
data1.index=list('ABCEDF')
data1
```

```python
#Calculate the coefficient of variation (CV = standard deviation / mean) for each row 
CV = {'CV':[
    (data1.iloc[0].std() / data1.iloc[0].mean()),
    (data1.iloc[1].std() / data1.iloc[1].mean()),
    (data1.iloc[2].std() / data1.iloc[2].mean()),
    (data1.iloc[3].std() / data1.iloc[3].mean()),
    (data1.iloc[4].std() / data1.iloc[4].mean()),
    (data1.iloc[5].std() / data1.iloc[5].mean())]}

CVdf=pd.DataFrame(CV, columns=['CV'])
CVdf.index=list('ABCEDF')

CVdf
```

```python
#clean up indeces and columns names
data1.reset_index(drop=True, inplace=True)
CVdf.reset_index(drop=True, inplace=True)

#update data1 by adding CV as a new column named ‘CV’.
data1=pd.concat([data1, CVdf], axis=1)
data1.index=list('ABCEDF')
data1
```

```python
# Calculate the mean of each column
Meandf = pd.DataFrame([data1.mean(axis=0)],["mean"], columns=('One', 'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'IQR', 'CV'))
Meandf
```

```python
#clean up indeces and columns names
data1.reset_index(drop=True, inplace=True)
Meandf.reset_index(drop=True, inplace=True)

#add mean as a new row named ‘mean’.
data1=data1.append(Meandf, ignore_index=True)
data1.index=['A', 'B', 'C', 'D', 'E', 'F', 'mean']

#print data1
data1
```
