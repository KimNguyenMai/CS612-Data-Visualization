```python
# In this exercise, you are asked to work on the API in the third section of HOP5 (from number 13). 
# You can access json data from the following url: http://jsonplaceholder.typicode.com/users.

# You can use the library ‘requests’ as you did in HOP5.

# From the user data, make a DataFrame df that consists of 
# ‘username’, ‘email’, ‘zipcode’, ‘lat’, ‘lng’, and ‘company_name’. 
# (‘zipcode’ is under ‘address’, ‘lat’, and ‘lng’ are under ‘geo’ under ‘address’, 
# and ‘company_name’ is ‘name’ under ‘company’.)

# Select the rows whose ‘lng’ values are greater than 0 and whose ‘lat’ is less than 0. 
# You may display all the columns. Name it df_a.

# Write an output file of df_a in the csv format named ‘output.csv’.
```

```python
import pandas as pd
import requests
```

```python
url='http://jsonplaceholder.typicode.com/users'
res= requests.get(url)
```

```python
all_users=res.json()
all_users
```

```python
#get all zipcodes and put into a list
zipcode = [all_users[0]['address']['zipcode'],
          all_users[1]['address']['zipcode'],
          all_users[2]['address']['zipcode'],
          all_users[3]['address']['zipcode'],
          all_users[4]['address']['zipcode'],
          all_users[5]['address']['zipcode'],
          all_users[6]['address']['zipcode'],
          all_users[7]['address']['zipcode'],
          all_users[8]['address']['zipcode'],
          all_users[9]['address']['zipcode']]
print(zipcode)
```

```python
#get all lats and put into a list
lat = [all_users[0]['address']['geo']['lat'],
          all_users[1]['address']['geo']['lat'],
          all_users[2]['address']['geo']['lat'],
          all_users[3]['address']['geo']['lat'],
          all_users[4]['address']['geo']['lat'],
          all_users[5]['address']['geo']['lat'],
          all_users[6]['address']['geo']['lat'],
          all_users[7]['address']['geo']['lat'],
          all_users[8]['address']['geo']['lat'],
          all_users[9]['address']['geo']['lat']]
print(lat)
```

```python
#get all lngs and put into a list
lng = [all_users[0]['address']['geo']['lng'],
          all_users[1]['address']['geo']['lng'],
          all_users[2]['address']['geo']['lng'],
          all_users[3]['address']['geo']['lng'],
          all_users[4]['address']['geo']['lng'],
          all_users[5]['address']['geo']['lng'],
          all_users[6]['address']['geo']['lng'],
          all_users[7]['address']['geo']['lng'],
          all_users[8]['address']['geo']['lng'],
          all_users[9]['address']['geo']['lng']]
print(lng)
```

```python
#get all company names and put into a list
companyname = [all_users[0]['company']['name'],
          all_users[1]['company']['name'],
          all_users[2]['company']['name'],
          all_users[3]['company']['name'],
          all_users[4]['company']['name'],
          all_users[5]['company']['name'],
          all_users[6]['company']['name'],
          all_users[7]['company']['name'],
          all_users[8]['company']['name'],
          all_users[9]['company']['name']]
print(companyname)
```

```python
# From the user data, make a DataFrame df that consists of 
# ‘username’, ‘email’, ‘zipcode’, ‘lat’, ‘lng’, and ‘company_name’. 
df=pd.DataFrame(all_users, columns=['username', 'email'])
df.loc[:,'zipcode'] = zipcode
df.loc[:,'lat'] = lat
df.loc[:,'lng'] = lng
df.loc[:,'company name'] = companyname
df
```

```python
#turn values of lat and lng columns from string to numeric types
df[["lat", "lng"]] = df[["lat", "lng"]].apply(pd.to_numeric)
```

```python
#Select the rows whose ‘lng’ values are greater than 0 and whose ‘lat’ is less than 0. 
# You may display all the columns. Name it df_a.

df_a = df.loc[(df["lng"] > 0) & (df["lat"] < 0)]
df_a
```

```python
# Write an output file of df_a in the csv format named ‘output.csv’.
df_a.to_csv('output.csv')
```
