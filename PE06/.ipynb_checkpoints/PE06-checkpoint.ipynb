```python
import pandas as pd
```

```python
sales = pd.read_csv('../../CS612-Summer-2020/IN/Kim-Nguyen/Module5/sales_data_sample.csv', encoding='latin')
```

```python
# 1.Convert the entire sales data into pandas DataFrame and name it data.
data = pd.DataFrame(sales)
data
```

```python
# 2. Count the number of ‘USA’ in ‘COUNTRY’ column.
data[(data['COUNTRY'] == 'USA')].count()['COUNTRY']
```

```python
#3. Delete all the rows whose ‘TERRITORY’ attribute is equal to ‘NaN’ using dropna().
data['TERRITORY'].dropna()
```

```python
#4. Count the number of ‘USA’ in ‘COUNTRY’ column again. 
data[(data['COUNTRY'] == 'USA')].count()['COUNTRY']
```

```python
#5. In data, filter duplicates based on the ‘ORDERNUMBER’ column using drop_duplicates().
# Keep the last row by passing keep=’last’. (Removing Duplicates in 7.2)
data.drop_duplicates(subset= ['ORDERNUMBER'], keep='last')
```

```python
# 6. Select all the rows where ‘PRODUCTLINE’ column is equal to ‘Motorcycles’.
data[data['PRODUCTLINE'] == 'Motorcycles']
```

```python
#7. Select the ‘QUANTITYORDERED’ column only and save it as qty. qty now becomes a Series.
qty = data.QUANTITYORDERED
qty
```

```python
# 8. Use cut to divide the qty data into four bins of an equal length. (Discretization and Binning in 7.2)
bins = pd.cut(qty, 4)
bins
```

```python
# 9. Use pd.value_counts() to count the number of data points in each bin in #8. 
pd.value_counts(bins)
```

```python
# 10. Use qcut to divide the qty data into four bins pertaining to each quartile. 
pd.qcut(qty,4)
```

```python
# 11. Use pd.value_counts() to count the number of data points in each bin in #10. 
pd.value_counts(pd.qcut(qty,4))
```
